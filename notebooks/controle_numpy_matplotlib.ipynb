# Partie E — Visualisations Avancées

### E22. Subplots comparatifs
Deux scatter plots côte à côte : Setosa vs Versicolor.


```python
# Votre code ici
setosa = features[species == 'Iris-setosa']
versicolor = features[species == 'Iris-versicolor']

SepalLength_setosa = setosa[:,0]
SepalWidth_setosa = setosa[:,1]

SepalLength_versicolor = versicolor[:,0]
SepalWidth_versicolor = versicolor[:,1]

PetalLength_setosa = setosa[:,2]
PetalWidth_setosa = setosa[:,3]

PetalLength_versicolor = versicolor[:,2]
PetalWidth_versicolor = versicolor[:,3]

fig , ax = plt.subplots(1,2, figsize=(10,6))

ax[0].scatter(SepalLength_setosa , SepalWidth_setosa , color='blue' , ec='yellow' , lw=2 , alpha=0.7 , label='setosa')
ax[0].scatter(SepalLength_versicolor , SepalWidth_versicolor , color='red' , ec='black' , lw=2 , alpha=0.7 , label='versicolor')
ax[0].set_xlabel('Sepal Length')
ax[0].set_ylabel('Sepal width')
ax[0].set_title('Sepal length vs width' , fontsize=14 , fontweight='bold')
ax[0].legend()
ax[0].grid(True , alpha=0.6)

ax[1].scatter(PetalLength_setosa , PetalWidth_setosa , color='blue' , ec='yellow' , lw=2 , alpha=0.7 , label='setosa' )
ax[1].scatter(PetalLength_versicolor , PetalWidth_versicolor , color='red' , ec='black' , lw=2 , alpha=0.7 , label='versicolor')
ax[1].set_xlabel('petal length')
ax[1].set_ylabel('petal width')
ax[1].set_title('petal length vs width')
ax[1].legend()
ax[1].grid(True , alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# Votre code ici
SepalLength1 = features[:,0]
SepalWidth1 = features[:,1]
PetalLength1 = features[:,2]
PetalWidth1 = features[:,3]

plt.subplot(2,2,1)
plt.hist(SepalLength1 , bins=15 , color='blue' , ec='black', lw=2)
plt.xlabel('Sepal Length')
plt.ylabel('Fréquence')
plt.title('histogramme  Sepal Length' , fontsize=14 , fontweight='bold')
plt.grid(True , alpha=0.3)

plt.subplot(2,2,2)
plt.hist(SepalWidth1 , bins=15 , color='yellow' , ec='black', lw=2)
plt.xlabel('Sepal Width')
plt.ylabel('Fréquence')
plt.title('histogramme  Sepal width' , fontsize=14 , fontweight='bold')
plt.grid(True , alpha=0.3)

plt.subplot(2,2,3)
plt.hist(PetalLength1 , bins=15 , color='red' , ec='black', lw=2)
plt.xlabel('petal Length')
plt.ylabel('Fréquence')
plt.title('histogramme  petal Length' , fontsize=14 , fontweight='bold')
plt.grid(True , alpha=0.3)

plt.subplot(2,2,4)
plt.hist(PetalWidth1 , bins=15 , color='skyblue' , ec='black', lw=2)
plt.xlabel('petal width')
plt.ylabel('Fréquence')
plt.title('histogramme  petal width' , fontsize=14 , fontweight='bold')
plt.grid(True , alpha=0.3)

plt.tight_layout()
plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
setosa = features[species == 'Iris-setosa']
versicolor = features[species == 'Iris-versicolor']
virginica = features[species == 'Iris-virginica']

petal_setosa = setosa[:,2]
petal_versicolor = versicolor[:,2]
petal_virginica = virginica[:,2]

moyennes = [np.mean(petal_setosa),np.mean(petal_versicolor), np.mean(petal_virginica)]

ecarts_type1 = [np.std(petal_setosa) , np.std(petal_versicolor) , np.std(petal_virginica)]

label = ['setosa' , 'versicolor' , 'virginica'  ]
colors2 = ['blue' , 'orange' , 'green']

plt.figure(figsize=(10,6))
plt.bar(label , moyennes , yerr=ecarts_type1 ,  capsize=8 ,color=colors2 , ec='black' , lw=2 )
plt.grid(True , axis='y' , alpha=0.6)

plt.xlabel('espèces')
plt.ylabel('moyenne de petal length')
plt.title('moyennes par espèce avec écart-type')

plt.show()
```

### E25. Scatter coloré par espèce
Scatter plot avec couleurs par espèce.


```python
# Votre code ici
setosa = features[species == 'Iris-setosa']
versicolor = features[species == 'Iris-versicolor']
virginica = features[species == 'Iris-virginica']

petal_length_setosa = setosa[:,2]
PetalWidth_setosa = setosa[:,3]

petal_length_versicolor = versicolor[:,2]
petalWidth_versicolor = versicolor[:,3]

petal_length_virginica = virginica[:,2]
petalWidth_virginica = virginica[:,3]

fig , axs = plt.subplots(1,1,figsize=(10,6))

axs.scatter(petal_length_setosa ,PetalWidth_setosa , c='blue' , ec='yellow' , lw=2, label='setosa' , alpha=0.7)
axs.scatter(petal_length_versicolor ,petalWidth_versicolor , c='red' , ec='black' , lw=2, label='versicolor' , alpha=0.7)
axs.scatter(petal_length_virginica ,petalWidth_virginica , c='purple' , ec='blue' , lw=2, label='virginica' , alpha=0.7)
axs.set_xlabel('petal Length')
axs.set_ylabel('petal width')
axs.set_title('Scatter plot avec couleurs par espèce' , fontsize=14 , fontweight='bold')
axs.legend()
axs.grid(True , alpha=0.4)

plt.tight_layout()
plt.show()
```

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
# Votre code ici
setosa = features[species == 'Iris-setosa']
versicolor = features[species == 'Iris-versicolor']
virginica = features[species == 'Iris-virginica']

PetalWidth_setosa_cumsum = np.cumsum(setosa[:,3])

petalWidth_versicolor_cumsum = np.cumsum(versicolor[:,3])

petalWidth_virginica_cumsum = np.cumsum(virginica[:,3])

label2 = ['setosa' , 'versicolor' , 'virginica'  ]

fig , ax2 = plt.subplots(1,1,figsize=(10,6))

ax2.plot(PetalWidth_setosa_cumsum , c='blue' ,marker='o', mec='red' , mfc='blue' , lw=2 , label='setosa' , alpha=0.7)
ax2.plot(petalWidth_versicolor_cumsum , c='orange' ,marker='o',  mec='blue' , mfc='orange' , lw=2 , label='versicolor' , alpha=0.7)
ax2.plot(petalWidth_virginica_cumsum , c='green' ,marker='o', mec='skyblue' , mfc='green' , lw=2 , label='virginica' , alpha=0.7)
ax2.set_xlabel('index des fleurs')
ax2.set_ylabel('cumsum de la largeur des pétales')
ax2.set_title('Cumsum des largeurs des pétales par espèce', fontsize=14 , fontweight='bold')
ax2.legend()
ax2.grid(True , alpha=0.6)

plt.tight_layout()
plt.show()
```
